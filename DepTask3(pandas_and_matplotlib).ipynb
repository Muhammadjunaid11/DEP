```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enabling inline plotting
%matplotlib inline

# Loading the dataset into a pandas DataFrame
titanic_data = pd.read_csv('/content/titanic.csv')

# Displaying the first few rows of the dataset
print(titanic_data.head())

# Displaying basic information about the dataset
print(titanic_data.info())

# Displaying summary statistics
print(titanic_data.describe())

# Displaying the count of each unique value in the 'Survived' column
print(titanic_data['Survived'].value_counts())

# Checking for missing values
print(titanic_data.isnull().sum())

# Filling missing values in the 'Age' column with the median age
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

# Filling missing values in the 'Embarked' column with the most common port
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

# Droping the 'Cabin' column due to a large number of missing values
titanic_data.drop(columns=['Cabin'], inplace=True)

# Verifying that there are no more missing values
print(titanic_data.isnull().sum())

# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by class
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Class')
plt.show()

# Age distribution of survivors and non-survivors
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='Age', hue='Survived', multiple='stack')
plt.title('Age Distribution of Survivors and Non-Survivors')
plt.show()

# Survival rate by embarkation port
sns.barplot(x='Embarked', y='Survived', data=titanic_data)
plt.title('Survival Rate by Embarkation Port')
plt.show()

# Fare distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```
